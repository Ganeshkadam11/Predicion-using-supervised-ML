```python
# Importing required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Loading the dataset
df =pd.read_csv("gterrorism.csv",encoding='ISO-8859-1') 
```

```python
df.head(5)
```

```python
df.columns
```

```python
#Checking the missing data
df.isnull().sum()
```

```python
df2=df.copy()
```

```python
df2.head()
```

```python
#Dropping the columns which has all null value
df2.dropna(axis=1, how='all')
```

```python
df2.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
df2 = df2[['Year','Month','Day','Country','state','Region','city','latitude','longitude','AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
df2['Wounded'] = df2['Wounded'].fillna(0).astype(int)
df2['Killed'] = df2['Killed'].fillna(0).astype(int) #fiilling the null value witn zero
```

```python
#Adding a new column 'casuality'
df2['casualities']=df2['Killed']+df2['Wounded']
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
print("The country with highest terror attack:",df2['Country'].value_counts().index[0])
print("Regions with highest terrorist attacks: ",df2['Region'].value_counts().index[0])
print("Highest people killed in an attack",df2['Killed'].max(),"that too place in",df2.loc[df2['Killed'].idxmax()])
print("City with the highest number of attacks:",df2['city'].value_counts().index[1])
print("Year with the most attacks:",df2['Year'].value_counts().idxmax())
print("Month with the most attacks:",df2['Month'].value_counts().idxmax())
print("Group with the most attacks:",df2['Group'].value_counts().index[1])
```

```python
print("The country with highest terror attack:",df2['Country'].value_counts().index[0])
print("Regions with highest terrorist attacks: ",df2['Region'].value_counts().index[0])
print("Highest people killed in an attack",df2['Killed'].max(),"that too place in",df2.loc[df2['Killed'].idxmax()].Country)
print("City with the highest number of attacks:",df2['city'].value_counts().index[1])
print("Year with the most attacks:",df2['Year'].value_counts().idxmax())
print("Month with the most attacks:",df2['Month'].value_counts().idxmax())
print("Group with the most attacks:",df2['Group'].value_counts().index[1])
print("Most Attack Types:",df2['AttackType'].value_counts().idxmax())
```

```python
f,ax = plt.subplots(figsize=(8,8))
sns.heatmap(df2.corr(),annot=False, linewidths=.4, fmt=".1f=",cmap=sns.diverging_palette(20,220,n=200), ax=ax)
```

The below plot shows the number of terrorist activities occure in each year

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df2, palette='coolwarm',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation= 90)
plt.title("Number of Terrorist Activity each Year")
plt.show()
```

From the above graph I can say that in the year 2014 maximum terrorist activities occures.

```python
#The below plot shows the most used attacking method by the terrorist
plt.subplots(figsize=(15,6))
sns.countplot('AttackType',data=df2,palette='rocket',order=df2['AttackType'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Attacking methods by terrorists")
plt.show()
```

From the above plot we can say that Bombing/Explosion is the most used attacking method by terrorists

```python
#The below plot shows the favorite targets of terrorists
plt.subplots(figsize=(15,6))
sns.countplot('Target_type',data=df2,palette='mako',order=df2['Target_type'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Favorite Targets")
plt.show()
```

From the above plot we can say private citizens and property are most targeted by terrorists

```python
#The below plot shows the most affected region by terrorists activity
sns.countplot('Region',data=df2,palette='Paired',edgecolor=sns.color_palette('dark',7),order=df2['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Number of Terrorist Activities By Region")
plt.show()
```

From the above plot we can say that Middle East and North Africa is the most affected region by terrorism.

```python
terror_region=pd.crosstab(df2.Year,df2.Region)
terror_region.plot(color=sns.color_palette('Paired',12))
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

```python
pd.crosstab(df2.Region,df2.AttackType).plot.barh(stacked=True,width=1,color=sns.color_palette('Paired',9))
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
#The below plot shows the most affected country
plt.subplots(figsize=(18,6))
sns.barplot(df2['Country'].value_counts()[:15].index,df2['Country'].value_counts()[:15].values,palette='rainbow')
plt.title('Top Affected Countries')
plt.show()
```

From the above plot we can say, Iraq is the most affected country by terrorism.

```python
#The below plot shows the number of attacks and how much people are wounded and killed
f,ax = plt.subplots(figsize=(18,10))
df2.Killed.plot(kind="line",color = "r", label="number of kills",linewidth=5, alpha=0.5, grid=True, linestyle=":")
df2.Wounded.plot(color="b", label="number of wounds",linewidth=5, alpha=0.5, grid=True,linestyle=":")
plt.legend(loc="upper right")
plt.xlabel("Number of attacks", size=15)
plt.ylabel("People", size=15)
plt.title("Line Plot")
```

```python

#The below plot shows the number of attacks that took place and the no. of people killed
coun_terror=df2['Country'].value_counts()[:15].to_frame()
coun_terror.columns=['Attacks']
coun_kill=df2.groupby('Country')['Killed'].sum().to_frame()
coun_terror.merge(coun_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```

In the above plot we can see that maximum attacks and killings took place in iraq

```python
#The below plot shows the no. of killings that took place over the years and what type of weapon was most used
sns.lmplot(x='Year', y='Killed',data=df2,fit_reg=False,hue='Weapon_type',legend=True ,                     # its subcategory   
           palette="Set1",aspect=15/10)
plt.xlabel("year")
plt.ylabel("kills") 
plt.title("Year vs Kill Scatter Plot")
```

```python
#The below plot shows the terrorist groups with highest terror attack
sns.barplot(df2['Group'].value_counts()[1:15].values,df2['Group'].value_counts()[1:15].index,palette=('inferno'))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Terrorist Organization with Highest Terror Attacks')
plt.show()
```

From the above plot we can say highest terrorist attacks are caused by Taliban terrorist organization
